# **Tugas 1**

```python
class QueueBioskop:
    def __init__(self, kapasitas=10):
        self.queue = []
        self.kapasitas = kapasitas
        self.total_pelanggan = 0

    def enqueue(self, tiket_id):
        if "2" in tiket_id:
            print(f"Pelanggan {tiket_id} langsung mendapat kursi tanpa antre!")
            self.total_pelanggan += 1
        else:
            if len(self.queue) >= self.kapasitas:
                print(f"Kapasitas bioskop penuh! Harap tunggu sampai ada kursi kosong.")
                return

            self.queue.append(tiket_id)
            print(f"Pelanggan {tiket_id} masuk antrean.")
            self.total_pelanggan += 1

    def dequeue(self):
        if not self.is_empty():
            pelanggan = self.queue.pop(0)
            self.total_pelanggan -= 1
            print(f"Pelanggan {pelanggan} dipanggil untuk membeli tiket.")
        else:
            print("Antrean kosong!")

    def display(self):
        print(f"Total pelanggan di bioskop: {self.total_pelanggan}/{self.kapasitas} (VIP tidak dihitung antrean)")
        if self.is_empty():
            print("Antrean kosong.")
        else:
            print("Antrean saat ini:", " -> ".join(self.queue))

    def is_empty(self):
        return len(self.queue) == 0


# Program utama
bioskop = QueueBioskop(kapasitas=10)

while True:
    print("\nMenu:")
    print("1. Tambah pelanggan ke antrean")
    print("2. Layani pelanggan")
    print("3. Lihat pelanggan pertama")
    print("4. Lihat pelanggan terakhir")
    print("5. Cek apakah antrean kosong")
    print("6. Lihat jumlah pelanggan dalam antrean")
    print("7. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        nama = input("Masukkan Nama Pelanggan: ")
        npm = input("Masukkan NPM Pelanggan: ")
        tiket_id = f"{nama[:2].upper()}{npm[-2:]}"
        bioskop.enqueue(tiket_id)
    elif pilihan == "2":
        bioskop.dequeue()
    elif pilihan == "3":
        print("Pelanggan pertama:", bioskop.queue[0] if not bioskop.is_empty() else "Kosong")
    elif pilihan == "4":
        print("Pelanggan terakhir:", bioskop.queue[-1] if not bioskop.is_empty() else "Kosong")
    elif pilihan == "5":
        print("Antrean kosong?", bioskop.is_empty())
    elif pilihan == "6":
        bioskop.display()
    elif pilihan == "7":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid. Coba lagi.")
```

# **Tugas 2**

```python
class FotokopiQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} diproses langsung (Fast Track) - {num_pages} halaman.")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} ditambahkan ke antrian - {num_pages} halaman.")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan yang diproses.")
            return None
        order_id, num_pages = self.queue.pop(0)
        if num_pages > 100:
            remaining_pages = num_pages - 100
            self.queue.insert(0, (order_id, remaining_pages))
            print(f"Pesanan {order_id} diproses 100 halaman, sisa {remaining_pages} halaman dalam antrian.")
        else:
            print(f"Pesanan {order_id} selesai diproses - {num_pages} halaman.")
        return order_id

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Program utama
antrian = FotokopiQueue()

while True:
        print("\nMenu:")
        print("1. Tambah Pesanan")
        print("2. Proses Pesanan")
        print("3. Lihat Antrian Pertama")
        print("4. Lihat Antrian Terakhir")
        print("5. Cek Apakah Antrian Kosong")
        print("6. Lihat Jumlah Pesanan")
        print("7. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            nama = input("Masukkan nama mahasiswa: ")[:3].upper()
            npm = input("Masukkan NPM mahasiswa: ")[-2:]
            order_id = f"{nama[:2].upper()}{npm[-2:]}"
            num_pages = int(input("Masukkan jumlah halaman: "))
            antrian.enqueue(order_id, num_pages)
        elif pilihan == "2":
            antrian.dequeue()
        elif pilihan == "3":
            print("Pesanan pertama:", antrian.front())
        elif pilihan == "4":
            print("Pesanan terakhir:", antrian.rear())
        elif pilihan == "5":
            print("Antrian kosong?", antrian.is_empty())
        elif pilihan == "6":
            print("Jumlah pesanan dalam antrian:", antrian.size())
        elif pilihan == "7":
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid, silakan coba lagi.")
```

# **Tugas 3**

```python
class ParkirQueue:
    def __init__(self, kapasitas=15):
        self.queue = []
        self.kapasitas = kapasitas

    def enqueue(self, vehicle_number):
        if len(self.queue) >= self.kapasitas:
            print("Parkiran penuh! Harap tunggu hingga ada kendaraan keluar.")
            return

        self.queue.append(vehicle_number)
        print(f"Kendaraan {vehicle_number} masuk antrean parkir.")

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan parkir.")
            return None
        vehicle_number = self.queue.pop(0)
        print(f"Kendaraan {vehicle_number} keluar dari parkiran.")
        return vehicle_number

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Program Utama
parkiran = ParkirQueue()

while True:
    print("\nMenu:")
    print("1. Tambah Kendaraan")
    print("2. Proses Kendaraan")
    print("3. Lihat Antrian Pertama")
    print("4. Lihat Antrian Terakhir")
    print("5. Cek Apakah Antrian Kosong")
    print("6. Lihat Jumlah Kendaraan")
    print("7. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        warna_plat = input("Masukkan warna plat kendaraan: ").strip().lower()
        kota = input("Masukkan Kota Lahir: ")[:2].upper()
        npm = input("Masukkan NPM: ")[-2:]
        vehicle_number = f"{kota}{npm}"

        if warna_plat == "biru":
            print(f"Parkir Prioritas! Kendaraan {vehicle_number} langsung masuk.")
        else:
            parkiran.enqueue(vehicle_number)

    elif pilihan == "2":
        parkiran.dequeue()

    elif pilihan == "3":
        print("Kendaraan Pertama:", parkiran.front())

    elif pilihan == "4":
        print("Kendaraan Terakhir:", parkiran.rear())

    elif pilihan == "5":
        print("Antrian Kosong?", parkiran.is_empty())

    elif pilihan == "6":
        print("Jumlah kendaraan dalam parkiran:", parkiran.size())

    elif pilihan == "7":
        print("Keluar dari program.")
        break

    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```

# **Tugas 4**

```python
class SertifikatQueue:
    def __init__(self):
        self.queue = []
        self.verified_codes = {"ABC123", "XYZ789", "LMN456"}

    def enqueue(self, student_id):
        self.queue.append(student_id)
        print(f"Mahasiswa {student_id} masuk antrean.")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada mahasiswa yang dilayani.")
            return None
        student_id = self.queue.pop(0)
        print(f"Mahasiswa {student_id} mengambil sertifikat.")
        return student_id

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Program utama
antrian = SertifikatQueue()

while True:
    print("\nMenu:")
    print("1. Tambah Mahasiswa")
    print("2. Layani Mahasiswa")
    print("3. Lihat Mahasiswa Pertama")
    print("4. Lihat Mahasiswa Terakhir")
    print("5. Cek Apakah Antrian Kosong")
    print("6. Lihat Jumlah Mahasiswa")
    print("7. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        prodi = input("Masukkan program studi: ")[:1].upper()
        npm = input("Masukkan NPM: ")[:2].upper()
        student_id = f"{prodi}{npm}"

        ikut_lomba = input("Apakah Anda pernah ikut lomba nasional? (ya/tidak): ").strip().lower()
        if ikut_lomba == "ya":
            kode_verifikasi = input("Masukkan kode verifikasi: ").strip()
            if kode_verifikasi in antrian.verified_codes:
                print(f"Mahasiswa {student_id} langsung mengambil sertifikat tanpa antrean!")
                continue  # Langsung selesai tanpa masuk antrean
            else:
                print("Kode verifikasi salah! Anda masuk antrean reguler.")

        antrian.enqueue(student_id)

    elif pilihan == "2":
        antrian.dequeue()
    elif pilihan == "3":
        print("Mahasiswa pertama dalam antrean:", antrian.front())
    elif pilihan == "4":
        print("Mahasiswa terakhir dalam antrean:", antrian.rear())
    elif pilihan == "5":
        print("Antrian kosong?", antrian.is_empty())
    elif pilihan == "6":
        print("Jumlah mahasiswa dalam antrean:", antrian.size())
    elif pilihan == "7":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```
