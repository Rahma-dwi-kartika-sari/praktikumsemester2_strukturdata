# **TUGAS**

```python
import time

class TowerOfHanoi:
    def __init__(self, n, max_steps):
        self.n = n
        self.max_steps = max_steps
        self.steps = 0
        self.towers = {'A': list(range(n, 0, -1)), 'B': [], 'C': []}

    def move(self, from_tower, to_tower):
        if self.steps >= self.max_steps:
            return
        self.towers[to_tower].append(self.towers[from_tower].pop())
        self.steps += 1
        print(f"\nLangkah {self.steps}: {from_tower} → {to_tower}")
        self.show()
        time.sleep(1)

    def solve(self, n, from_tower, to_tower, aux_tower):
        if self.steps >= self.max_steps:
            return
        if n == 1:
            self.move(from_tower, to_tower)
        else:
            self.solve(n - 1, from_tower, aux_tower, to_tower)
            self.move(from_tower, to_tower)
            self.solve(n - 1, aux_tower, to_tower, from_tower)

    def show(self):
        """Menampilkan kondisi tower dalam format vertikal dengan susunan yang benar"""
        max_height = self.n
        tower_A, tower_B, tower_C = self.towers['A'], self.towers['B'], self.towers['C']

        A_list = (["|"] * (max_height - len(tower_A))) + tower_A[::-1]
        B_list = (["|"] * (max_height - len(tower_B))) + tower_B[::-1]
        C_list = (["|"] * (max_height - len(tower_C))) + tower_C[::-1]

        print("\n         KEADAAN MENARA")
        print("     ======================")
        for i in range(max_height):
            print(f"       {A_list[i]:^3}    {B_list[i]:^3}    {C_list[i]:^3}  ")
        print("     ----------------------")
        print("        A      B      C    \n")

# Implementasi dengan npm akhir 8
npm = 8
hanoi = TowerOfHanoi(n=3, max_steps=npm)

print("\nKEADAAN AWAL:")
hanoi.show()

hanoi.solve(3, 'A', 'C', 'B')

print("\nKEADAAN AKHIR:")
hanoi.show()
```
