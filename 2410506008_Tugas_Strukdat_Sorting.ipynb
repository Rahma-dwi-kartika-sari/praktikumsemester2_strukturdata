```python
def quick_sort(data, key):
    if len(data) <= 1:
        return data

    pivot = data[len(data) // 2]
    left = [x for x in data if x[key] < pivot[key]]
    middle = [x for x in data if x[key] == pivot[key]]
    right = [x for x in data if x[key] > pivot[key]]

    return quick_sort(left, key) + middle + quick_sort(right, key)


# Data pelanggan
customers = [
    {"name": "Ada Wong", "age": 24, "total_spent": 123500},
    {"name": "Albert", "age": 38, "total_spent": 432700},
    {"name": "Claire", "age": 25, "total_spent": 543000},
    {"name": "Chris", "age": 19, "total_spent": 1243999},
    {"name": "Leon", "age": 21, "total_spent": 700000}
]

# Menampilkan data sebelum sorting
print("\n=== Data Pelanggan Sebelum Sorting ===")
print(f"{'Nama':<10} {'Umur':<5} {'Total Belanja':<10}")
print("=" * 30)
for customer in customers:
    print(f"{customer['name']:<10} {customer['age']:<5} {customer['total_spent']:<10}")

while True:
    # Menu pilihan
    print("\n=== Menu Pengurutan Data Pelanggan ===")
    print("1. Nama (A-Z)")
    print("2. Umur (Termuda ke Tertua)")
    print("3. Total Belanja (Terkecil ke Terbesar)")
    print("4. Selesai (Keluar)")

    choice = input("Masukkan pilihan (1/2/3/4): ")

    if choice == "1":
        key = "name"
    elif choice == "2":
        key = "age"
    elif choice == "3":
        key = "total_spent"
    elif choice == "4":
        print("Program selesai. Terima kasih!")
        break
    else:
        print("Pilihan tidak valid, coba lagi.")
        continue

    # Sorting dengan Quick Sort
    sorted_customers = quick_sort(customers[:], key)

    # Menampilkan output setelah sorting
    print("\n=== Data Setelah Sorting ===")
    print(f"{'Nama':<10} {'Umur':<5} {'Total Belanja':<10}")
    print("=" * 30)
    for customer in sorted_customers:
        print(f"{customer['name']:<10} {customer['age']:<5} {customer['total_spent']:<10}")
```
